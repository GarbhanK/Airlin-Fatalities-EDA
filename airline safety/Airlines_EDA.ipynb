```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv('airline-safety.csv')
```

```python
# Specify the data
airlines = df['airline']
FA_85_99 = df['fatal_accidents_85_99']



# PLot it out
figure(figsize=(24, 10), dpi=100)
plt.bar(airlines, FA_85_99.sort_values())
plt.xticks(rotation=45)
plt.title("Fatal Accidents between 1985 and 1999")
plt.show()
```

```python
# Specify the data
airlines = df['airline']
FA_00_14 = df['fatal_accidents_00_14']



# PLot it out
figure(figsize=(24, 10), dpi=100)
plt.bar(airlines, FA_85_99.sort_values(), alpha=0.5, color='blue')
plt.bar(airlines, FA_00_14.sort_values(), alpha=0.5, color='red')
plt.xticks(rotation=45)
plt.title("Fatal Accidents between 85-99 VS. 00-14")
plt.show()
```
